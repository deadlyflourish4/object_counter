```python
from ultralytics import YOLO

# Load the YOLO26 model
model = YOLO("yolo26m.pt")

# Export the model to ONNX format
model.export(format="onnx", opset=21)  # creates 'yolo26m.onnx'

# Load the exported ONNX model
onnx_model = YOLO("yolo26m.onnx")

# Run inference on CPU
results = onnx_model("1.jpg", device="cpu")
```

```python

import onnxruntime as ort
sess = ort.InferenceSession("yolo26m.onnx")
print("Inputs:", [i.name for i in sess.get_inputs()])
print("Outputs:", [o.name for o in sess.get_outputs()])
```

```python
# Visualize the results
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np

# Get the first result
result = results[0]

# Plot the result
result.plot()
plt.show()

# Display image with detections
im_array = result.plot(conf=True)  # plot predictions with confidence scores
im = Image.fromarray(im_array[..., ::-1])
im.show()
```
